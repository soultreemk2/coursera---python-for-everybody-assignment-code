```python
# 가장 기본적인 그래프
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,100) # x축 정의. 0~10을 100개로 나눔
plt.plot(x,np.sin(x))  # 사인 커브를 그린것
plt.show() # 그림의 끝에 입력(안써도 그림은 나오지만 습관처럼 써주기)
```

```python
# 두개의 그래프를 하나에 그리기
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.show()
```

```python
# x축, y축 범위 지정
plt.plot(x,np.sin(x))
plt.xlim(-1,11) # x축 범위 (폭 지정)
plt.ylim(-1.5,1.5) # y축 범위
plt.show()
```

```python
plt.plot(x,np.sin(x))
plt.axis([-1,11,-1.5,1.5]) # x, y축 범위 한번에 정함
plt.show()
```

```python
# 그래프 스타일 변경
# matplotlib에 설정되어 있는 그래프의 스타일에 관련해서는 다음의 링크 참고
# https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html  
plt.plot(x,np.sin(x),'b-o') # 파란색(b) 동그라미(o)로 표시한 후 줄(-)로 연결 
plt.show()
```

```python
plt.style.use('seaborn-white')
plt.plot(x,np.sin(x),'b',label='sin')
plt.plot(x,np.cos(x),label='cos')
plt.title('sine-cosine curve')
plt.legend() # 범례(label준 거를 표시해줌)
plt.show()
```

```python
plt.plot(x,np.sin(x),label='sin')
plt.plot(x,np.cos(x),label='cos')
plt.legend(loc='upper right') # 범례 위치를 오른쪽 위로 설정
plt.xlabel('x')
plt.ylabel('sine and cosine')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
ax=plt.axes() # ax를 통해서 똑같이 할 수 있음(axes를 통해 접근하겠다.)
ax.plot(x,np.sin(x),'k-') # 검은색 줄로 연결
ax.plot(x,np.cos(x))
plt.show()
```

```python
ax=plt.axes()
ax.plot(x,np.sin(x))
ax.set(xlim=(0,10),ylim=(-2,2),xlabel='x',ylabel='sun(x)',title='sine curve') 
# 범위, 라벨, 제목 한번에 설정
# axes의 set을 통해 한번에 때려넣을수 있음(교수님은 개인적으로 안좋아한다함)
plt.show()
```

```python
# 2개 이상의 그래프를 하나로 묶어서 출력
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,100)
plt.figure() # 시작 (그림을 따로따로 떨어뜨려서 그리고싶을때)

plt.subplot(2,1,1) # subplot(행, 열, 몇번째 plot인지) # 그림이 2x1 matrix형태로 들어감. 그 중 1번째 그림
plt.plot(x,np.sin(x),'b--',label='sine')
plt.plot(x,np.cos(x),label='cosine')
plt.title('sine and cosine curve')

plt.subplot(2,1,2)
plt.plot(x,np.cos(x))

plt.show() # 끝
```

```python
fig, ax = plt.subplots(2,1)
ax[0].plot(x,np.sin(x),'b--',label='sine')
ax[0].plot(x,np.cos(x),label='cosine')
ax[0].set_title('sine and cosine curve')
ax[0].legend(loc='best')
ax[1].plot(x,np.cos(x))
plt.show()
```

```python
fig, ax = plt.subplots(2,2,figsize=(8,5)) # 2x2 그림에 대한 크기를 조정해줄 수 있음
ax[0][0].plot([11,22,33,44],'ro--') # 첫번째 행과 첫번째 열
ax[0][1].plot([[1,2],[3,4]],[[11,22],[33,44]],'o-') # 첫번째 행과 두번째 열
ax[1][0].plot(np.linspace(0,5),np.cos(2*np.pi*np.linspace(0,5))) # 두번째 행의 첫번째 열
ax[1][1].plot([3,5],[3,5],'bo:') # 두번째 행과 두번째 열
ax[1][1].plot([3,5],[5,3],'kx--')
plt.show()
### ax를 쓰면 각각의 그림의 위치를 control할 수 있음 
```

```python
# seaborn을 이용한 그래프 그리기
import matplotlib.pyplot as plt
import seaborn as sns
sns.set() 
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x),'b--',label='sine')
plt.legend(loc='best')
plt.show()
```

```python
# 히스토그램 그리기
import pandas as pd
bn=np.random.multivariate_normal([0,0],[[5,1],[1,2]],size=1000) # 평균이 0, x1과x2의 분산의 5,2 / 1은 covariance
bn=pd.DataFrame(bn,columns=['x','y'])
print(bn)
for i in 'xy':
    plt.hist(bn[i],density=True,alpha=0.2)  

# X,Y를 하나씩 읽으면서 히스토그램을 그려줌
# density=True는 y를 frequency가 아니고 백분율로 바꿔줌
# alpha는 그림의 투명도인데 1에 가까울수록 불투명
```

```python
# Kernel density estimation(kde)를 이용하여 그리기
for i in 'xy':
    sns.kdeplot(bn[i],shade=True) # 적은 데이터로도 부드러운 곡선 형태로 그리는법

sns.distplot(bn['x']) 
```

```python
sns.distplot(bn['y'])
```

```python
sns.kdeplot(bn) # kde 추정치
```

```python
# kde와 주변분포 그리기
sns.jointplot(bn['x'],bn['y'],kind='kde')
```

```python
# 회귀선과 주변분포 그리기
sns.jointplot(x='x',y='y',data=bn,kind='reg')
```

```python
# 각 pair에 대한 산포도 그리기
iris=sns.load_dataset('iris')
iris.head()
```

This is very useful for **exploring correlations between multidimensional
data, when you’d like to plot all pairs of values against each other.**

```python
sns.pairplot(iris,hue='species') # hue란 문자화되있는 변수를 색깔별로 구분
# 이런 그림을 쭉 그려주면 어떤 데이터를 선택할지 쉽게 파악할 수 잇음
```

```python
# 그룹별 분포 알아보기
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.factorplot(x='day',y='total_bill',hue='sex',data=tips,
               row='smoker',col='time',kind='box') # smoker와 time 조합별 boxplot
```

```python
grid=sns.FacetGrid(data=tips,row='smoker',col='sex',margin_titles=True) # 열과 행에 그룹 정보 표시
grid.map(plt.hist,'tip')
```
